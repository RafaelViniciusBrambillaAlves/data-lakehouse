```python
from processing.spark.utils.spark_session import create_spark_session
```

```python
spark = create_spark_session("explore-gold")
```

### gold_ratings_over_time

Mostra como os ratings evoluem ao longo do tempo. Permite identificar períodos com maior volume de avaliações e variações na média de notas e sentimento dos usuários.

```python
gold_ratings_over_time = spark.read.format("parquet").load("s3a://lakehouse/gold.db/gold_ratings_over_time")
```

```python
gold_ratings_over_time.limit(10).show(truncate = False)
```
